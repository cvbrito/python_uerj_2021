**Exercício 1 - Maratonista com velocidade (quase) constante**

```python
import matplotlib.pyplot as plt

vel = 12*(60/(3.6)) # velocidade em m/min
d = 5000 # distância em metros
tempo = d/vel

lista_pos = []
lista_tempo = []

for i in range(int(tempo)+1):
  pos = vel*i
  lista_pos.append(pos)
  lista_tempo.append(i)

print(lista_pos)
print(lista_tempo)

plt.plot(lista_tempo, lista_pos)
plt.xlabel("Tempo (min)")
plt.ylabel("Distância (m)")
plt.show()
```

**Exercício 2 - Gráfico movimento acelerado**

```python
import math
import matplotlib.pyplot as plt

# primeira parte: movimento retilíneo uniforme

vel_inicial = 12/(3.6) # em m/s
dist_mru1 = 5000 # por uma questão de conveniência, desprezamos o restante do movimento
tempo_mru1 = int(dist_mru1/vel_inicial) # tempo (em segundos), que dura o movimento

lista_pos = [] # lista com as posições do atleta
lista_tempo = [] # lista com os tempos das medições de posição

for t in range(int(tempo_mru1)+1):
  pos_mru1 = vel_inicial*t
  lista_pos.append(pos_mru1)
  lista_tempo.append(t)
  tempo_inicial_mrua = lista_tempo[t]

# segunda parte: movimento retilíneo uniformemente acelerado

vel_final = 15/(3.6) # velocidade após a acelração, em m/s
delta_s = 200 # distância em metros percorida durante a aceleração, em m
acel = ((vel_final)**2-(vel_inicial)**2)/(2*delta_s) # em m/(s**2)
tempo_acel = int((vel_final-vel_inicial)/acel) # tempo que dura a aceleração

for i in range(int(tempo_acel)+1):
  if i == 0:
    i = i+1
  pos_mrua = dist_mru1 + vel_inicial*i + acel*(i**2)/2
  delta_t = i + tempo_inicial_mrua
  lista_pos.append(pos_mrua)
  lista_tempo.append(delta_t)
  tempo_inicial_mru2 = lista_tempo[len(lista_tempo)-1]
  pos_inicial_mru2 = lista_pos[len(lista_pos)-1]

# terceira parte: movimento retilíneo uniforme

dist_mru2 = 1800
tempo_mru2 = int(dist_mru2/vel_final) # tempo (em segundos) que dura a terceira fase do movimento


for i in range(int(tempo_mru2)+1):
  pos_mru2 = pos_inicial_mru2 + vel_final*i
  delta_t2 = i + tempo_inicial_mru2
  lista_pos.append(pos_mru2)
  lista_tempo.append(delta_t2)


plt.plot(lista_tempo, lista_pos)
plt.xlabel("Tempo (s)")
plt.ylabel("Distância (m)")
plt.show()
```

**3. Pêndulo simples em um plano**

```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# inserindo valores

g = 9.81 # m/s**2
l = 1 # m
m = 0.2 # kg
w = math.sqrt(g/l)

def equation(y, t):
  theta, x = y
  dydt = [x, -(w**2)*math.sin(theta)]
  return dydt

# condições iniciais

y0 = [-math.pi/2, 0]

# intervalos de tempos

tempo = np.linspace(0, 10, 201)

sols = odeint(equation, y0, tempo)
'''
plt.plot(tempo, sols[:, 0])
plt.legend(loc='lower right')
plt.ylabel('Ângulo (rad)')
plt.xlabel('Tempo (t)')
plt.show()
'''

# calculando x(t) e y(t)

lista_x = []
lista_y = []

for i in sols[:,0]:
  x_t = l*math.sin(i)
  y_t = l*(1-math.cos(i))
  lista_x.append(x_t)
  lista_y.append(y_t)

array_x = np.array(lista_x)
array_y = np.array(lista_y)

plt.plot(tempo, array_x, label = 'x(t)')
plt.plot(tempo, array_y, label = 'y(t)')
plt.legend(loc='best')
plt.ylabel('Posição (m)')
plt.xlabel('Tempo (t)')
plt.show()
```
