**Exercício 1 - Desenhar flores**

```python
import turtle, math, random

tart = turtle.Turtle()
tart.ht()
tart.speed(0)

def poly(t, angle, n_tracos, tam_traco):
  i = 1
  while i <= (angle / (360/n_tracos)):
    t.fd(tam_traco)
    t.lt(360/n_tracos)
    i = i + 1

def arc(t, r, angle):
  comprimento_arco = angle * 2 * math.pi * r / 360
  comprimento_circ = 2 * math.pi * r
  n_tracos = 200
  tam_traco = (comprimento_circ / n_tracos)
  poly(t, angle, n_tracos, tam_traco)

def random_color():
#    R = random.randrange(0, 255, 10)
    B = random.randrange(0, 255, 10)
    G = random.randrange(0, 255, 10)

    tart.color(200, G, B)

def petala(t, r):
  ''' Desenha uma petala usando um arco de 60 graus e raio r '''
  heading = t.heading()
  t.begin_fill()
  arc(t, r, 60)
  t.lt(180-60)
  arc(t, r, 60)
  t.end_fill()
  t.seth(heading)
  
def flor(t, r, n):
  ''' Desenha uma flor com n pétalas que são arcos de 60 graus e raio r '''
  for i in range(n):
    random_color()
    petala(t, r)
    t.lt(360/n)
  
def caule(t, r):
#  Desenha o caule e a folha de uma flor
  t.color("green")
  t.goto(0,-2*r)
  t.seth(90)
  t.fd(r/2)
  t.seth(0)
  petala(t, r/2)
  t.seth(90)
  t.fd(3*r/2)

def flor_completa(t, r, n):
# Desenha uma flor completa, com talo, folhas e pétalas.
  caule(t, r)
  flor(t, r, n)
  
flor_completa(tart, 70, 12)
```

**Exercício 2 — Desenhar um polígono e suas linhas internas**

```python
import turtle, math

tart = turtle.Turtle()
tart.ht()
tart.speed(0)

def lines(t, length, n):
  alpha = 360/n
  beta = (180-alpha)/2
  d = (length)/(2*math.sin(math.pi/n))
  
  for i in range(n):
    t.fd(length)
    t.lt(alpha+beta)
    t.fd(d)
    t.lt(180)
    t.fd(d)
    t.lt(180-beta)
    
lines(tart, 30, 25)
```

**Exercício 5 - Desenhando uma espiral**

```python
import turtle, math

tart = turtle.Turtle()
tart.speed(0)
tart.ht()

def espiral(t, n_tracos):
  a = 1
  b = 5
  for i in range(n_tracos):
    theta = i/(b**2)*math.pi
    x = (a+b*theta)*math.cos(theta) # equação em coordenadas cartesianas
    y = (a+b*theta)*math.sin(theta)
    t.goto(x,y)

espiral(tart, n_tracos = 200)
```
